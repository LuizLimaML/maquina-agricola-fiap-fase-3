## 1. Imports e carregamento dos dados

## 4. Pre-processamento

```python
le = LabelEncoder()
y = le.fit_transform(df['cultura'])
X = df[['umidade', 'fosforo', 'potassio', 'ph', 'temperatura']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## 5. Treinar 5 modelos

```python
modelos = {
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'RandomForest': RandomForestClassifier(random_state=42),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC()
}

resultados = []
for nome, m in modelos.items():
    m.fit(X_train, y_train)
    pred = m.predict(X_test)
    acc = accuracy_score(y_test, pred)
    resultados.append({'modelo': nome, 'accuracy': acc})

pd.DataFrame(resultados).sort_values('accuracy', ascending=False)
```

## 6. Conclusao
Os modelos foram testados em dados sinteticos, entao a accuracy nao reflete um caso real. O dataset eh pequeno (50 linhas) e as features sao geradas aleatoriamente, sem padrao real entre cultura e as variaveis. Em producao seria necessario coletar dados de verdade e fazer feature engineering. O RandomForest e a Logistic Regression costumam ser bons baselines.
